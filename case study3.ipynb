```python
%matplotlib inline
import pandas as pd
import numpy as np
import statistics as st
import seaborn as sn
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\HP\Desktop\Sales_add.csv')
```

```python

#np.DataFrame=df
np.DataFrame=df
list(df.columns.values)
# null hyothesis : no significant difference in sales after marketing
#alt hypothesis: there is significant difference in sales after marketing
bfsales=df[df['Sales_before_digital_add(in $)']>0]
afsales=df[df['Sales_After_digital_add(in $)']>0]
sn.distplot(bfsales['Sales_before_digital_add(in $)'],kde=False,label='sales bf marketing')
sn.distplot(afsales['Sales_After_digital_add(in $)'],kde=False,label='sales af marketing')
plt.legend()
plt.show()
# as we can observe from the graph sales after marketing is increased so fail to accept null hypothesis ie alternate hypothesis
```
